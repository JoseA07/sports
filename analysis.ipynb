```python
team_ids = {}
team_counter = 1
```

```python
def get_team_id(team):
    global team_counter
    if team not in team_ids:
        team_ids[team] = f"T{team_counter:05}"  # Ensure IDs are at least 3 characters
        team_counter += 1
    return team_ids[team]

# Apply the function to create IDs
df['home_team_id'] = df['home_team'].apply(get_team_id)
df['away_team_id'] = df['away_team'].apply(get_team_id)
```
