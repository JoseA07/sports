```python
def fetch_data(league: str, headers: dict = None) -> pd.DataFrame:
    """
    Get data from an API:
    - get the datafrom football api

    Parameters:
    - url (str): api url
    - headers (dict): adict with params for the dict

    Returns:
    - pd.DataFrame: DataFrame football data
    """
    url = f'https://api.football-data.org/v4/competitions/{league}/matches?status=FINISHED'
    response = requests.get(url, headers=headers)
    data = response.json()['matches']
    df = pd.json_normalize(data)
    df = df.drop(columns=['group', 'odds.msg'])
    return df
```

```python
def standardize_columns(df: pd.DataFrame) -> pd.DataFrame:
    """
    Standardize column names in a DataFrame:
    - Converts camelCase to snake_case
    - Replaces dots (.) and spaces with underscores (_)
    - Converts to lowercase and removes special characters

    Parameters:
    - df (pd.DataFrame): Input DataFrame

    Returns:
    - pd.DataFrame: DataFrame with standardized column names
    """
    df.columns = [
        re.sub(r'([a-z])([A-Z])', r'\1_\2', col)   # Convert camelCase to snake_case
        .replace('.', '_')                         # Replace dots with underscores
        .replace(' ', '_')                         # Replace spaces with underscores
        .lower()                                   # Convert to lowercase
        for col in df.columns
    ]
    return df
```

```python
def to_snake_case(name):
    """Convert to snake case column name"""
    name = re.sub("(.)([A-Z][a-z]+)", r"\1_\2", name)
    name = re.sub("__([A-Z])", r"_\1", name)
    name = re.sub("([a-z0-9])([A-Z])", r"\1_\2", name)
    name = name.replace('.', '_')                         
    name = name.replace(' ', '_') 
    return name.lower()
```

```python
def get_scorers_data(league: str, headers: dict = None) -> pd.DataFrame:
    """
    Get data from an API:
    - get the datafrom football api

    Parameters:
    - url (str): api url
    - headers (dict): adict with params for the dict

    Returns:
    - pd.DataFrame: DataFrame football data
    """
    url = f'https://api.football-data.org/v4/competitions/{league}/scorers'
    response = requests.get(url, headers=headers)
    data = response.json()["scorers"]
    df = pd.json_normalize(data)
    return df
```
